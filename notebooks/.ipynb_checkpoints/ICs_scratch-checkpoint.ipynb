```python
def get_surface_profile(sn, PartType=0, center=np.array([200, 200, 200]), field='Masses'):
    part = getattr(sn, 'part0')
    pos = np.subtract(part.pos, center)
    mass = getattr(part, field)
    
    R = np.linalg.norm(pos[:,:2], axis=1)
    
    aveR_list = []
    surf_list = []
    
    Rbins = np.logspace(-1, 1.2, 15)
    for i in range(len(Rbins)-1):
        Rmin = Rbins[i]
        Rmax = Rbins[i+1]
        
        keys = np.where(np.logical_and(R > Rmin, R < Rmax))[0]
        totmass = np.sum(mass[keys])
        totarea = np.pi * (Rmax**2 - Rmin**2)
        
        aveR = np.mean(R[keys])
        
        surf = totmass / totarea
#         surf *= (1E10 / 1E6) # convert to Msun/pc^2
        
        aveR_list.append(aveR)
        surf_list.append(surf)
    
    return np.array(aveR_list), np.array(surf_list)
        
```

```python
aveR_old, surf_old = get_surface_profile(snap_old)
aveR_new, surf_new = get_surface_profile(snap_new)
```

```python
fig, ax = plt.subplots(1, 1)

ax.plot(aveR_old, surf_old)
ax.plot(aveR_new, surf_new)

ax.set(yscale='log')

plt.show()
```

```python
def get_surface_map(snap, prop, xextent, Nxbins, yextent=None, Nybins=None, center=np.array([200, 200, 200])):
    if yextent is None:
        yextent = xextent
    if Nybins is None:
        Nybins = Nxbins
    
    bins = (Nxbins, Nybins)
    
    pos = np.subtract(snap.part0.pos, center)
    x = pos[:,0]
    y = pos[:,1]
    
    xlen = (xextent[1] - xextent[0]) / Nxbins
    ylen = (yextent[1] - yextent[0]) / Nybins
    
    weights = np.array(getattr(snap.part0, prop)) / (xlen * ylen)
#     print(weights)
    
    rnge = [[xextent[0], xextent[1]], [yextent[0], yextent[1]]]
    print(rnge)
#     print(bins)
    
    heatmap, xdg, ydg = np.histogram2d(pos[:,0], pos[:,1], bins=bins, weights=weights, range=rnge)
    
    extent = [xdg[0], xdg[1], ydg[0], ydg[1]]
    
    return heatmap, extent
    
```
